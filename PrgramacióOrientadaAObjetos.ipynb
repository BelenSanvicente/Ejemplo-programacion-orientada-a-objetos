```python
# Ana Belen Sanvicente Méndez

listaEstudiantes = []                                                                                                       # Lista vacía

class Estudiantes(object):                                                                                                  # Nombre de la clase
    def __init__(self, _cedula, _nombre, _apellido, _edad, _nota1, _nota2, _nota3):                                         # Propiedades o carceterísticas de la clase. Init es el método constructor
        self.cedula = _cedula                                                                                               # Definición de las variables de instancia
        self.nombre = _nombre
        self.apellido = _apellido
        self.edad = _edad
        self.nota1 = _nota1
        self.nota2 = _nota2
        self.nota3 = _nota3
        self.notaFinal = (_nota1 + _nota2 + _nota3) / 3
        self.historial = []
    def entregarDatos(self):                                                                                               # Función para entregar datos
        print("No. Cedula: {} - {} {} - Nota Final: {}".format(self.cedula, self.nombre, self.apellido, self.notaFinal))   # Mostrar o imprimirlos datos, y formato del orden en que aparecerán los datos dentro del string
    def editarNotas(self, _nota1, _nota2, _nota3):                                                                         # Función para editar las notas con 3 variables de intancia 
        self.nota1 = _nota1                                                                                                # Defincición de las variables de instancia
        self.nota2 = _nota2
        self.nota3 = _nota3
        self.notaFinal = (_nota1 + _nota2 + _nota3) / 3
        print("Registro Exitoso!")
    def incluirEvento(self, _nota1, _nota2, _nota3):                                                                      # Función con la palabra reservada self y 3 variables para las notas
        return ("modificacion: Nota_1: {} Nota_2: {} Nota_3: {}".format(_nota1, _nota2, _nota3))                          # FInal de la función e indicar el valor de la función en un string
    def entregaHistorial(self):                                                                                           # Función
        print("No. Cedula: {} - {} {}".format(self.cedula, self.nombre, self.apellido))


def registrarEstudiante():                                                                                                # Función para registrar los datos del estudiante
    print("Registro de Estudiantes\n")                                                                                    # Mensaje de instrucción en la consola
    cedula = int(input("Ingrese el numero de cedula: "))                                                                  # Entrada para que ingrese la cédula tipo entero
    nombre = input("Ingrese el nombre: ")                                                                                 # Mensaje para ingresar el nombre del estudiante   
    apellido = input("Ingrese el apellido: ")                                                                             # Y el apellido
    edad = int(input("Ingrese su edad: "))                                                                                # Mensaje para ingresar la edad de tipo entero
    nota1 = float(input("Ingrese nota 1: "))                                                                              # Mensaje para ingresar las calificaciones de tipo float
    nota2 = float(input("Ingrese nota 2: "))
    nota3 = float(input("Ingrese nota 3: "))
    objAlumno = Estudiantes(cedula, nombre, apellido, edad, nota1, nota2, nota3)                                          # Nueva variable para crear la lista
    listaEstudiantes.append(objAlumno)                                                                                    # Llenado de la lista con los datos ingresados a partir de la variable definida en la celda anterior

def listadoEstudiantes():                                                                                                # Función 
    print("Listado de Estudiantes\n")                                                                                    # Mostrar el título de la lista como mensaje en la consola
    for objAlumno in listaEstudiantes:                                                                                   # Inicio del ciclo for para que nos entregue los datos de cada estudiante
        objAlumno.entregarDatos()                                                                                        # Acción y fin del ciclo for
                                                                        
def buscarEstudiante():                                                                                                 # Función para buscar un estudiante con la cédula y para que muestre los datos de dicho estudiante
    print("Buscar Estudiante\n")
    cedula = int(input("Ingrese el numero de cedula a buscar: "))
    for objAlumno in listaEstudiantes:
        if cedula == objAlumno.cedula:
            objAlumno.entregarDatos()

def modificarNotas():                                                                                                   # Función para cambiar las calificaciones de tipo float de un alumno en particular
    print("Modificar Notas\n")
    cedula = int(input("Ingrese el numero de cedula a buscar: "))
    for objAlumno in listaEstudiantes:
        if cedula == objAlumno.cedula:
            nota1 = float(input("Ingrese nota 1: "))                                                                    # Nueva calificación 1
            nota2 = float(input("Ingrese nota 2: "))                                                                    # Nueva calificación 2
            nota3 = float(input("Ingrese nota 3: "))                                                                    # Nueva calificación 3
            objAlumno.editarNotas(nota1, nota2, nota3)
            objAlumno.entregarDatos()
            recepcionMensaje = objAlumno.incluirEvento(nota1, nota2, nota3)
            objAlumno.historial.append(recepcionMensaje)

def consultarHistorial():                                                                                               # Función para consultar el historial
    print("Consulta de Historial\n")                                                                                    # Mensaje en la consola (título)
    cedula = int(input("Ingrese el numero de cedula a buscar: "))                                                       # Cedula del esdiante del que se requiere consultar
    for objAlumno in listaEstudiantes:                                                                                  # Inicio del ciclo for
        if cedula == objAlumno.cedula:                                                                                  # Condicional de si la cedula está definida 
            for recepcionMensaje in objAlumno.historial:                                                                # Inicio del ciclo for 
                print("Evento: {}".format(recepcionMensaje))                                                            # IMprime el historial del alumno identificado ppor la cédula

def salir():                                                                                                            # Función para salir
# Acciones de la función
    print("Salida inminente...!")                                                                                       # Mensaje de un string de la salida
    exit()                                                                                                              # palabra reservada exit

def main():                                                                                                             # Función para el menú
    while True:
        print("\n")                                                                                                     # Imprimir salto de linea
        print("|****************************|")                                                                         # Mensaje de bienvanida
        print("|**|      Bienvenidos     |**|")
        print("|**|         Menu         |**|")
        print("|****************************|")
        print("")                                                                                                      # Espacio por medio de un string
        print("Seleccione una de las siguientes opciones:");                                                           # Mensaje de intrucción a seguir
        print("1.- Registrar Estudiante")                                                                              # Mensaje en la consola con un string para la Opcion 1
        print("2.- Mostrar Estudiantes")                                                                               # Mensaje en la consola con un string para la Opcion 2
        print("3.- Buscar Estudiante")                                                                                 # Mensaje en la consola con un string para la Opcion 3
        print("4.- Modificar notas")                                                                                   # Mensaje en la consola con un string para la Opcion 4
        print("5.- Consultar Historial")                                                                               # Mensaje en la consola con un string para la Opcion 5
        print("6.- Salir\n")                                                                                           # Mensaje en la consola con un string para la Opcion 6

        opcion = int(input("Opcion: "))

        if opcion == 1:                                                                                               # Condicional principal, la opcion 1 corresponde a la función registrar estudiante.
            registrarEstudiante()
        elif opcion == 2:                                                                                             # Uso de elif para que solo se ejecute una opcion de las siguientes, la opción 2 corresponde a la función listado de estudiantes.
            listadoEstudiantes()
        elif opcion == 3:                                                                                             # Opción 3 corresponde a la función de buscar estudiante.
            buscarEstudiante()
        elif opcion == 4:                                                                                             # Opción 4 corresponde a la función de modificar la calificación de estudiante.
            modificarNotas()
        elif opcion == 5:                                                                                             # Opción 5 corresponde a la función de consultar historial.
            consultarHistorial()
        elif opcion == 6:                                                                                             # Opción 6 corresponde a salir.
            salir()

if __name__ == '__main__':                                                                                            # Inicio del programa, Condicional para ejecutar la funcion main.
    main(); 
```
